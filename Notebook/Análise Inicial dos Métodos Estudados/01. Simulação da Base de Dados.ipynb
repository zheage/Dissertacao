```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gerando a base de dados com 20 covariáveis e 1% de eventos
X, y = make_classification(
    n_samples=5000, 
    n_features=2,
    n_informative=2, 
    n_redundant=0,
    n_clusters_per_class=2, 
    weights=[0.95], 
    flip_y = 0.005, # Fração dos dados que será atribuída aleatoriamente um rótulo diferente
    random_state=42
)


# Criando um DataFrame com as covariáveis
df_covariates = pd.DataFrame(X, columns=['Covariável 1', 'Covariável 2'])

# Adicionando a coluna de eventos
df_covariates['Evento'] = y

# Plotando o gráfico de dispersão
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_covariates, x='Covariável 1', y='Covariável 2', hue='Evento', palette=['steelblue', 'crimson'])
plt.title('Dispersão das Covariáveis')
plt.xlabel('Covariável 1')
plt.ylabel('Covariável 2')

legend = plt.legend(title='Evento', loc='upper right')

plt.show()
```

```python
# Gerando a base de dados com 20 covariáveis e 1% de eventos
X, y = make_classification(
    n_samples=100000, 
    n_features=20,
    n_informative=10, 
    n_redundant=10, 
    n_clusters_per_class=3, 
    weights=[0.99], 
    flip_y = 0.005, # Fração dos dados que será atribuída aleatoriamente um rótulo diferente
    random_state=42
)

# Criando o DataFrame
df = pd.DataFrame(X, columns=[f'COL_{i+1}' for i in range(20)])
df['Evento'] = y

# Exibindo as primeiras linhas do DataFrame
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Contagem dos eventos
event_counts = df['Evento'].value_counts()

# Renomeando as labels
labels = ['Não-evento', 'Evento']

# Criação do gráfico de rosca
plt.figure(figsize=(8, 6))
plt.pie(event_counts, labels=labels, colors=['steelblue', 'crimson'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title('Distribuição da Variável Resposta')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Primeiro gráfico: Dispersão das Covariáveis
sns.scatterplot(data=df_covariates, x='Covariável 1', y='Covariável 2', hue='Evento', palette=['steelblue', 'crimson'], ax=axes[0])
axes[0].set_title('Dispersão das Covariáveis')
axes[0].set_xlabel('Covariável 1')
axes[0].set_ylabel('Covariável 2')
axes[0].legend(title='Evento', loc='upper right')

# Segundo gráfico: Distribuição da Variável Resposta
event_counts = df['Evento'].value_counts()
labels = ['Não-evento', 'Evento']
axes[1].pie(event_counts, labels=labels, colors=['steelblue', 'crimson'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
axes[1].set_title('Distribuição da Variável Resposta')

plt.tight_layout()
plt.show()
```
