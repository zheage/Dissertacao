```python
import pandas as pd
import numpy as np
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Gerando a base de dados com 20 covariveis e 1% de eventos
X, y = make_classification(
    n_samples=5000, 
    n_features=2,
    n_informative=2, 
    n_redundant=0,
    n_clusters_per_class=2, 
    weights=[0.95], 
    flip_y = 0.005, # Frao dos dados que ser atribuda aleatoriamente um rtulo diferente
    random_state=42
)


# Criando um DataFrame com as covariveis
df_covariates = pd.DataFrame(X, columns=['Covarivel 1', 'Covarivel 2'])

# Adicionando a coluna de eventos
df_covariates['Evento'] = y

# Plotando o grfico de disperso
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_covariates, x='Covarivel 1', y='Covarivel 2', hue='Evento', palette=['steelblue', 'crimson'])
plt.title('Disperso das Covariveis')
plt.xlabel('Covarivel 1')
plt.ylabel('Covarivel 2')

legend = plt.legend(title='Evento', loc='upper right')

plt.show()
```

```python
# Gerando a base de dados com 20 covariveis e 1% de eventos
X, y = make_classification(
    n_samples=100000, 
    n_features=20,
    n_informative=10, 
    n_redundant=10, 
    n_clusters_per_class=3, 
    weights=[0.99], 
    flip_y = 0.005, # Frao dos dados que ser atribuda aleatoriamente um rtulo diferente
    random_state=42
)

# Criando o DataFrame
df = pd.DataFrame(X, columns=[f'COL_{i+1}' for i in range(20)])
df['Evento'] = y

# Exibindo as primeiras linhas do DataFrame
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configurando o estilo do Seaborn
sns.set(style="whitegrid")

# Contagem dos eventos
event_counts = df['Evento'].value_counts()

# Renomeando as labels
labels = ['No-evento', 'Evento']

# Criao do grfico de rosca
plt.figure(figsize=(8, 6))
plt.pie(event_counts, labels=labels, colors=['steelblue', 'crimson'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
plt.title('Distribuio da Varivel Resposta')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Primeiro grfico: Disperso das Covariveis
sns.scatterplot(data=df_covariates, x='Covarivel 1', y='Covarivel 2', hue='Evento', palette=['steelblue', 'crimson'], ax=axes[0])
axes[0].set_title('Disperso das Covariveis')
axes[0].set_xlabel('Covarivel 1')
axes[0].set_ylabel('Covarivel 2')
axes[0].legend(title='Evento', loc='upper right')

# Segundo grfico: Distribuio da Varivel Resposta
event_counts = df['Evento'].value_counts()
labels = ['No-evento', 'Evento']
axes[1].pie(event_counts, labels=labels, colors=['steelblue', 'crimson'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
axes[1].set_title('Distribuio da Varivel Resposta')

plt.tight_layout()
plt.show()
```
