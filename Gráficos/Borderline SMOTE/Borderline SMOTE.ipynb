```python
import numpy as np
from imblearn.over_sampling import BorderlineSMOTE
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('dark')

# Gerar um conjunto de dados de exemplo
X, y = make_classification(n_classes=2, class_sep=2, weights=[0.9, 0.1],
                           n_informative=2, n_redundant=0, flip_y=0.05,
                           n_features=2, n_clusters_per_class=1,
                           n_samples=200, random_state=25)

# Aplicar Borderline-SMOTE
smote = BorderlineSMOTE(k_neighbors=5, m_neighbors=10, random_state=42, sampling_strategy = 0.33)
X_res, y_res = smote.fit_resample(X, y)

# Plotar os dados originais e os dados após Borderline-SMOTE
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Dados originais
ax[0].scatter(X[y == 0][:, 0], X[y == 0][:, 1], label='Classe Minoritária', color='steelblue')
ax[0].scatter(X[y == 1][:, 0], X[y == 1][:, 1], label='Classe Majoritária', color='crimson')
ax[0].set_title('Dados originais')
ax[0].set_xticks([])
ax[0].set_yticks([])

# Dados após Borderline-SMOTE
ax[1].scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label='Classe Minoritária', color='steelblue')
ax[1].scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label='Classe Majoritária', color='crimson')
ax[1].set_title('Dados após Borderline-SMOTE')
ax[1].set_xticks([])
ax[1].set_yticks([])

plt.show()
```
