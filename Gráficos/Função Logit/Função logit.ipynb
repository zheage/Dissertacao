```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo o estilo do gráfico
sns.set(style="whitegrid")

# Definindo a função logit
def logit(x):
    return 1 / (1 + np.exp(-x))

# Calculando os valores de y usando a função logit
y = logit(x)

# Criando o gráfico
plt.figure(figsize=(12, 5))
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('logit(x)')
plt.grid(True)
plt.show()
```
